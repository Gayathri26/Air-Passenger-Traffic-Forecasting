# Data preprocessing



1.   Drop unwanted columns
2.   Compute month column
2.   Rename columns
4.   Convert to Lowercase values
5.   Reorder the columns
6.   Split cities to remove state details




Dropping unwanted columns

```python
domestic_df  = domestic_df .drop(['UNIQUE_CARRIER_NAME','ORIGIN','DEST'], axis = 1) 
```

Compute month columns

```python
domestic_df['MONTH'] = domestic_df['MONTH'].apply(str)
```

```python
month_dict = {'1':'January',
              '2':'February',
              '3':'March',
              '4':'April',
              '5':'May',
              '6':'June',
              '7':'July',
              '8':'August',
              '9':'September',
              '10':'October',
              '11':'November',
              '12':'December'
            }
domestic_df['month_name'] = domestic_df.MONTH.map(month_dict)
```

Rename columns


```python
rename_dict = {'passengers count	':'passengers count','ORIGIN_CITY_NAME':'origin_city','ORIGIN_STATE_NM':'origin_state','DEST_CITY_NAME':'dest_city','DEST_STATE_NM':'dest_state','YEAR':'year','MONTH':'month_num','month_name':'month'}
domestic_df.rename(columns= rename_dict, inplace=True)
```

Convert to lowercase values

```python
domestic_df['origin_city'] = domestic_df['origin_city'].str.lower()
domestic_df['origin_state'] = domestic_df['origin_state'].str.lower()
domestic_df['dest_city'] = domestic_df['dest_city'].str.lower()
domestic_df['dest_state'] = domestic_df['dest_state'].str.lower()
domestic_df['month'] = domestic_df['month'].str.lower()
```

Reorder columns

```python
domestic_df = domestic_df[['year','month_num','month','origin_city','origin_state','dest_city','dest_state','passengers count']]
```

Splitting the cities columns based on ',' to remove state information

```python
domestic_df['origin_city'] = domestic_df['origin_city'].str.split("," , expand = True)
```

```python
domestic_df['dest_city'] = domestic_df['dest_city'].str.split("," , expand = True)
```

```python
domestic_df.shape
```

```python
domestic_df.to_csv('intermittent-result/Domestic US/domestic_preprocessed_main.csv')
```

# Seasonwise dataframes

```python
domestic_df.head()
```

Grouping based on months and summing the passenger counts

```python
agg_dict = {'month_num':'first', 'passengers count':'sum'}
domestic_grouped_df = domestic_df.groupby(['year','month','origin_city','dest_city'], as_index=False).agg(agg_dict)
```

```python
domestic_grouped_df.head()
```

All seasons dataframe

```python
#spring
domestic_spring_df = domestic_grouped_df[domestic_grouped_df.month.isin(['march', 'april', 'may'])]

#summer
domestic_summer_df = domestic_grouped_df[domestic_grouped_df.month.isin(['june', 'july', 'august'])]

#autumn
domestic_autumn_df = domestic_grouped_df[domestic_grouped_df.month.isin(['september', 'october', 'november'])]

#winter
domestic_winter_df = domestic_grouped_df[domestic_grouped_df.month.isin(['december', 'january', 'february'])]
```

```python
print('spring df size',domestic_spring_df.shape)
print('summer df size',domestic_summer_df.shape)
print('autumn df size',domestic_autumn_df.shape)
print('winter df size',domestic_winter_df.shape)
```

Creating csv files for 4 seasons

```python
domestic_spring_df.to_csv('intermittent-result/Domestic US/dom_spring.csv')
domestic_summer_df.to_csv('intermittent-result/Domestic US/dom_summer.csv')
domestic_autumn_df.to_csv('intermittent-result/Domestic US/dom_autumn.csv')
domestic_winter_df.to_csv('intermittent-result/Domestic US/dom_winter.csv')
```
